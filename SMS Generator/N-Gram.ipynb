```python
txt = ""
with open("./smsspamcollection//SMSSpamCollection.txt") as f:
    for line in f:
        txt += line
```

```python
import re

txt = re.sub(r"\bham\b","",txt)
```

```python
txt =re.sub(r"\bspam\b","",txt)
```

```python
from nltk import word_tokenize
tokens = word_tokenize(txt)
```

```python
tokens
```

```python
txt=" ".join(tokens)
import markovify
model = markovify.Text(txt)
for i in range(5):
    print(model.make_sentence())
```

```python
for i in range(5):
    print(model.make_short_sentence(100))
```

```python
unigram = markovify.Text(txt,state_size = 1)
for i in range(5):
    print(unigram.make_sentence_with_start("I"))
```

```python
bigram = markovify.Text(txt,state_size = 2)
for i in range(5):
    print(bigram.make_sentence_with_start("I"))
```
